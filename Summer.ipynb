## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df1 = pd.read_csv("summer.csv")

df1 = df1[df1['Medal']== 'Gold']

lst = [] 


for sport in df1['Sport'].unique():
    lst.append([sport, len(df1[df1['Sport'] == sport])])
    

lst = pd.DataFrame(lst, columns = ['Sport', 'No._of_gold_medal']).sort_values(by = 'No._of_gold_medal', ascending = False).head().plot(x = 'Sport', y = 'No._of_gold_medal', kind = 'bar', figsize = (5,5))

```
